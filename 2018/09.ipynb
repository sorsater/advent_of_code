```python
import re
from collections import defaultdict
from collections import deque
```

```python
data = open("input/09").read()
```

```python
re0 = r"(\d+) players; last marble is worth (\d+) points"
num_players, last_marble = list(map(int, re.match(re0, data).groups()))
```

```python
# First I went with a normal list but ofcourse took too long time in part2, switched to a deque
def run(num_players, num_marbles):
    scores = defaultdict(int)
    marbles = deque([0])
    cur_player = 1
    for mar in range(1, num_marbles + 1):
        if mar % 23 == 0:
            marbles.rotate(7)
            scores[cur_player] += mar + marbles.pop()
            marbles.rotate(-1)
        else:
            marbles.rotate(-1)
            marbles.append(mar)
        cur_player = (cur_player + 1) % num_players
    return max(scores.values())
```

```python
part1 = run(num_players, last_marble)
print(f"Answer #1: {part1}")
```

```python
part2 = run(num_players, last_marble * 100)
print(f"Answer #2: {part2}")
```
