```python
import networkx as nx
```

```python
data = open("input/25").read().splitlines()
```

```python
def create_graph(ignore_list=None):
    g = nx.Graph()
    added_nodes = set()
    for line in data:
        start_node, nodes = line.split(": ")
        if start_node not in added_nodes:
            g.add_node(start_node)

        for node in nodes.split(" "):
            if node not in added_nodes:
                g.add_node(node)
            
            for a, b in ignore_list:
                if a == start_node and b == node:
                    break
                if a == node and b == start_node:
                    break
            else:
                g.add_edge(start_node, node)
    nx.draw(g, with_labels=True)
    
    if ignore_list:
        subgraphs = nx.connected_components(g)
        prod = 1
        for s in subgraphs:
            prod *= len(s)
        print("Answer #1:", prod)   
```

```python
create_graph([])
```
