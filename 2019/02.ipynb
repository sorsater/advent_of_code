# Part 2

```python
def verb_noun():
    for verb in range(100):
        for noun in range(100):
            data = raw_data[:]
            data[1] = noun
            data[2] = verb
            c = IntCode(data)
            c.run()
            if c.program[0] == 19690720:
                return 100 * noun + verb
```

```python
part2 = verb_noun()
```

```python
print(f"Answer #2: {part2}")
```
