```python
from collections import defaultdict
from tqdm.notebook import tqdm
```

```python
data = open("input/24").read().splitlines()
```

```python
grid = {}
for r, line in enumerate(data):
    for c, elem in enumerate(line):
        grid[tuple((r, c))] = elem
```

```python
neighs = [[1, 0], [-1, 0], [0, 1], [0, -1]]
```

```python
def next_bug(pos):
    adjacent = 0
    for x, y in neighs:
        elem = grid.get(tuple((pos[0] + x, pos[1] + y)))
        if elem == "#":
            adjacent += 1
    
    return "." if adjacent != 1 else "#"
```

```python
def next_open(pos):
    adjacent = 0
    for x, y in neighs:
        elem = grid.get(tuple((pos[0] + x, pos[1] + y)))
        if elem == "#":
            adjacent += 1
    
    return "#" if 1 <= adjacent <= 2 else "."
```

```python
def iterate():
    new_grid = {}
    for pos, elem in grid.items():
        if elem == "#":
            new_grid[pos] = next_bug(pos)
        elif elem == ".":
            new_grid[pos] = next_open(pos)
    return new_grid
```

```python
def get_string_of_grid():
    out = ""
    for elem in grid.values():
        out += elem
    return out
```

```python
state_set = set()
while True:
    grid = iterate()
    state = get_string_of_grid()
    
    if state in state_set:
        break
    state_set.add(state)
```

```python
def calculate_score(grid):
    score = 0
    mult = -1
    for r in range(5):
        for c in range(5):
            mult += 1
            val = grid[tuple((r, c))]
            if val == ".":
                continue
            score += 2**mult

    return score
```

```python
part1 = calculate_score(grid)
```
