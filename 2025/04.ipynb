```python
data = open("input/04").read().splitlines()
```

```python
directions = [
    1,        # right
    -1,       # left
    1j,       # down
    -1j,      # up
    1 + 1j,   # down-right
    1 - 1j,   # up-right
    -1 + 1j,  # down-left
    -1 - 1j,  # up-left
]
```

```python
grid = set()
for r, row in enumerate(data):
    for c, elem in enumerate(row):
        if elem == "@":
            grid.add(r + c * 1j)
```

```python
initial_count = len(grid)
```

```python
def remove():
    mark_for_removal = set()
    for pos in grid:
        num_neighs = 0
        for n in directions:
            if pos + n in grid:
                num_neighs += 1
        if num_neighs < 4:
            mark_for_removal.add(pos)

    for pos in mark_for_removal:
        grid.remove(pos)
```

```python
remove()
```

```python
part1 = initial_count - len(grid)
assert part1 == 1569
```

```python
print(f"Answer #1: {part1}")
```

```python
prev_count = len(grid)
while True:
    remove()
    if (new_count := len(grid)) == prev_count:
        break
    prev_count = new_count
```

```python
part2 = initial_count - len(grid)
```

```python
assert part2 == 9280
```

```python
print(f"Answer #2: {part2}")
```
