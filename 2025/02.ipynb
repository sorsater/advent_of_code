```python
from tqdm.notebook import tqdm
```

```python
data = open("input/02").read()
```

```python
def is_invalid_part1(id):
    half = len(id) // 2
    return id[:half] == id[half:]
```

```python
def is_invalid_part2(id):
    if len(id) == 1:
        return False

    if len(set(id)) == 1:
        return True

    for step in range(2, len(id) // 2 + 1):
        if len(id) % step != 0:
            continue
        curr_chunk = id[:step]
        chunk_count = 0
        for idx in range(0, len(id), step):
            chunk = id[idx:idx + step]
            if chunk != curr_chunk:
                break
            chunk_count += 1
        else:
            if chunk_count >= 2:
                return True
    return False
```

```python
invalid_ids_part1 = 0
invalid_ids_part2 = 0
for r in tqdm(data.split(",")):
    start, end = list(map(int, r.split("-")))
    for id in range(start, end + 1):
        if is_invalid_part1(str(id)):
            invalid_ids_part1 += id
        if is_invalid_part2(str(id)):
            invalid_ids_part2 += id
```

```python
print(f"Answer #1: {invalid_ids_part1}")
```

```python
print(f"Answer #2: {invalid_ids_part2}")
```
