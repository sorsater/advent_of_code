```python
from collections import defaultdict
```

```python
data = open("input/12").read().splitlines()
```

```python
shapes = defaultdict(list)
rules = []
found_rules = False
cur_shape = 0
cur_shape_data = []
for line in data:
    if not line:
        shape_size = 0
        for row in cur_shape_data:
            shape_size += row.count("#")
        shapes[cur_shape] = [shape_size, cur_shape_data]
        cur_shape_data = []
        continue
    if "x" in line:
        found_rules = True
    
    if found_rules:
        size, items = line.split(": ")
        width, height = size.split("x")
        items = list(map(int, items.split(" ")))
        rules.append([int(width), int(height), items])
    else:
        if ":" in line:
            cur_shape = int(line.split(":")[0])
        else:
            cur_shape_data.append(line)
```

```python
# Read a comment that you should check the input before starting...
# I just checked to see if it was possible to fit them or not (enough space) and apparently that was enough...
```

```python
valid = 0
for width, height, items in rules:
    item_sizes = sum([item * shapes[idx][0] for idx, item in enumerate(items)])
    if item_sizes < width * height:
        valid += 1
```

```python
print(f"Answer #1: {valid}")
```
