```python
from collections import defaultdict
```

```python
data = open("input/01").read().splitlines()
```

```python
dial = 50
dials = defaultdict(int) 
dial_extra = 0

for line in data:
    right = 1 if line[0] == "R" else -1
    for _ in range(int(line[1:])):
        dial += right
        if dial % 100 == 0:
            dial_extra += 1

    dial %= 100

    # If it ends at 0, don't count the last "extra"
    if dial == 0:
        dial_extra -= 1

    dials[dial] += 1
```

```python
print(f"Answer #1: {dials[0]}")
```

```python
print(f"Answer #2: {dials[0] + dial_extra}")
```
