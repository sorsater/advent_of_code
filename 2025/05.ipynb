```python
from dataclasses import dataclass
```

```python
data = open("input/05").read().splitlines()
```

```python
@dataclass
class Range:
    first: int
    second: int

    def in_range(self, recipe):
        return self.first <= recipe <= self.second
```

```python
ranges = []
recipies = []
for line in data:
    if line == "":
        continue
    if "-" in line:
        ranges.append(Range(*list(map(int, line.split("-")))))
    else:
        recipies.append(int(line))
```

```python
def in_any_range(recipe):
    for r in ranges:
        if r.in_range(recipe):
            return 1
    return 0
```

```python
part1 = sum([in_any_range(recipe) for recipe in recipies]) 
```

```python
print(f"Answer #1: {part1}")
```

```python
def find_overlap(ranges):
    for idx1, r1 in enumerate(ranges):
        for idx2, r2 in enumerate(ranges):
            # First I checked "if r1.first == r2.first and r1.second == r2.second:"
            # But I had duplicate ranges...
            if idx1 == idx2:
                continue
                
            if r1.in_range(r2.first):
                return [r1, r2]
    return []
```

```python
def merge_ranges(ranges):
    merge_me =  find_overlap(ranges)
    if not merge_me:
        return ranges
        
    r1, r2 = merge_me

    # Create the new range
    new_ranges = [
        Range(min(r1.first, r2.first), max(r1.second, r2.second))
    ]
    # Add all the others
    for r in ranges:
        if r != r1 and r != r2:
            new_ranges.append(r)
        
    return new_ranges
```

```python
while True:
    new_ranges = merge_ranges(ranges)
    if len(ranges) == len(new_ranges):
        break
    ranges = new_ranges[:]
```

```python
part2 = sum([r.second - r.first + 1 for r in ranges])
```

```python
print(f"Answer #2: {part2}")
```
