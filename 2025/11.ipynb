```python
data = open("input/11").read().splitlines()
```

```python
connections = {}
for line in data:
    from_, to = line.split(": ")
    connections[from_] = to.split(" ")
```

```python
# DFS
def find_all_paths():
    q = ["you"]
    num_paths = 0
    while q:
        head = q.pop()
        if head == "out":
            num_paths += 1
            continue
        for elem in connections[head]:
            q.append(elem)
    return num_paths
```

```python
part1 = find_all_paths()
print(f"Answer #1: {find_all_paths()}")
assert part1 == 724
```

# Part 2
DFS was to heavy, using DFS with memoization

Realized I can just count the amount of paths for (assumin fft is before dac):
1. "svr" to "fft"
2. "fft" to "dac"
3. "dac" to "out"

and multiply them together.

In that way I don't need to check that the path contains the elements.

```python
def find_all_paths(start, goal, ignore, memo):
    if start in ignore:
        return 0

    if start == goal:
        return 1

    if start in memo:
        return memo[start]

    num_paths = 0
    for elem in connections[start]:
        num_paths += find_all_paths(elem, goal, ignore, memo)
    memo[start] = num_paths

    return num_paths
```

```python
# Decide if fft is before dac or other way around
```

```python
fft_first = find_all_paths("fft", "dac", {"out"}, {})
if fft_first > 0:
    a = find_all_paths("svr", "fft", {"dac", "out"}, {})
    b = find_all_paths("fft", "dac", {"out"}, {})
    c = find_all_paths("dac", "out", {}, {})
    part2 = a * b * c
else:
    a = find_all_paths("svr", "dac", {"fft", "out"}, {})
    b = find_all_paths("dac", "fft", {"out"}, {})
    c = find_all_paths("fft", "out", {}, {})
    part2 = a * b * c
```

```python
print(f"Answer #2: {part2}")
```

```python
assert part2 == 473930047491888
```
