```python
data = open("input/03").read().splitlines()
```

```python
def get_highest(line, length=2):
    candidates = list(map(int, list(line)))
    number = ""
    offset = 0
    for num in range(length):
        need_to_keep = length - num - 1
        if need_to_keep == 0:
            current_candidates = candidates[offset:]
        else:
            current_candidates = candidates[offset:-need_to_keep]

        max_val = max(current_candidates)
        number += str(max_val)
        
        max_idx = current_candidates.index(max_val)
        offset += max_idx + 1
        
    return int(number)
```

# Part 1

```python
tot = sum([get_highest(line, 2) for line in data])
print(f"Answer #1: {tot}")
```

# Part 2

```python
tot = sum([get_highest(line, 12) for line in data])
print(f"Answer #2: {tot}")
```
