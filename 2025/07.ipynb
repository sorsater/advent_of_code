```python
from collections import defaultdict
```

```python
data = open("input/07").read().splitlines()
```

```python
width = len(data[0])
height = len(data)
```

```python
grid = {}
for r, row in enumerate(data):
    for c, elem in enumerate(row):
        if elem == "S":
            start = (r, c)
            grid[tuple((r + 1, c))] = "|"
        if elem == "^":
            grid[tuple((r, c))] = elem
```

```python
def get_below(pos):
    return tuple((pos[0] + 1, pos[1]))

def get_below_left(pos):
    return tuple((pos[0] + 1, pos[1] - 1))

def get_below_right(pos):
    return tuple((pos[0] + 1, pos[1] + 1))

def get_above(pos):
    return tuple((pos[0] - 1, pos[1]))
```

```python
part1 = 0
for r in range(1, height):
    for c in range(width):
        pos = tuple((r, c))
        if grid.get(pos) != "|":
            continue
            
        below_pos = get_below(pos)
        below_elem = grid.get(below_pos)
        if below_elem == "^":
            part1 += 1
            grid[get_below_left(pos)] = "|"
            grid[get_below_right(pos)] = "|"
        else:
            grid[below_pos] = "|"
```

```python
print(f"Answer #1: {part1}")
```

Got inspiration from a GIF on AoC reddit to come up with this solution

```python
scores = defaultdict(int)
scores[tuple((start[0] + 1, start[1]))] = 1
for r in range(height):
    for c in range(width):
        pos = tuple((r, c))
        if (elem := grid.get(pos)) is None:
            continue
            
        above_pos = get_above(pos)
        if grid.get(above_pos) == "|":
            scores[pos] += scores[above_pos]
        if elem == "^":
            scores[get_below_left(pos)] += scores[pos] 
            scores[get_below_right(pos)] += scores[pos]
```

```python
part2 = 0
for pos, value in scores.items():
    if pos[0] == height - 1:
        part2 += value
```

```python
print(f"Answer #2: {part2}")
```
