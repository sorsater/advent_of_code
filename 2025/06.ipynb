```python
from functools import reduce
```

```python
data = open("input/06").read().splitlines()
```

```python
nums = []
ops = []
for idx, line in enumerate(data):
    line = [e for e in line.split(" ") if e != ""]
    if idx == len(data) - 1:
        ops = line
    else:
        nums.append(list(map(int, line)))  
```

```python
part1 = 0
for idx in range(len(ops)):
    op = ops[idx]
    tmp_nums = []
    for num in nums:
        tmp_nums.append(num[idx])
    if op == "+":
        part1 += sum(tmp_nums)
    else:
        part1 += reduce(lambda x, y: x * y, tmp_nums)
```

```python
print(f"Answer #1: {part1}")
```

# Part 2

```python
parts = []
part = []
for idx in range(len(data[0]) - 1, -1, -1):
    col = []
    for row in data[:-1]:
        col.append(row[idx])
    col = "".join(col)
    
    # Check for empty column
    maybe_empty = list(set(col))
    if len(maybe_empty) == 1 and maybe_empty[0] == " ":
        parts.append(part)
        part = []
    else:
        part.append(int(col))
parts.append(part)
```

```python
part2 = 0
op_idx = len(parts) - 1
for part in parts:
    op = ops[op_idx]
    if op == "+":
        part2 += sum(part)
    else:
        part2 += reduce(lambda x, y: x * y, part)
    op_idx -= 1
```

```python
print(f"Answer #2: {part2}")
```
