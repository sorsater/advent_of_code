```python
data = open("input/10").read().splitlines()
```

```python
wanted_indices = [20, 60, 100, 140, 180, 220]
```

```python
def get_crt():
    crt = []
    for row in range(6):
        line = []
        for col in range(40):
            line.append(".")
        crt.append(line)
    return crt
```

```python
class CRT:
    def __init__(self):
        self.crt = get_crt()
        self.x = 1
        self.tot_val = 0
        self.tick = 0
        
    def tick_me(self):
        self.draw()
        new_tick = self.tick + 1
        if new_tick in wanted_indices:
            self.tot_val += new_tick * self.x
        self.tick = new_tick
        

    def run_instr(self, instr):
        if instr[:4] == "noop":
            self.tick_me()
        else:
            value = int(instr.split("addx ")[1])
            self.tick_me()
            self.tick_me()
            self.x += value

    def draw(self):
        row, idx = divmod(self.tick, 40)
        if abs(idx - self.x) <= 1:
            self.crt[row][idx] = "#"
        
    def print(self):
        for row in self.crt:
            for col in row:
                if col == ".":
                    print(" ", end="")
                else:
                    print(col, end="")
            print()
```

```python
crt = CRT()
for instr in data:
    crt.run_instr(instr)
```

```python
print(f"Answer #1: {crt.tot_val}")
```
