```python
data = open("input/25").read().splitlines()
```

```python
char_map = {
    "1": 1,
    "2": 2,
    "0": 0,
    "-": -1,
    "=": -2,
}
```

```python
def snafu_to_dec(snafu):
    tot = 0
    for mult_idx, char in enumerate(snafu[::-1]):
        tot += 5**mult_idx * char_map[char]
    
    return tot
```

```python
tot = 0
for line in data:
    tot += snafu_to_dec(line)
```

```python
# Needed some help understanding divmod was the way to go. Was first trying to iteratorivel build it up myself..
def num_to_snafu(number):
    output = ""
    while number != 0:
        number, remainder = divmod(number, 5)
        if remainder <= 2:
            output += str(remainder)
        elif remainder == 3:
            output += "="
            number += 1
        else:
            output += "-"
            number += 1
    return output[::-1]
```

```python
part1 = num_to_snafu(tot)
```
